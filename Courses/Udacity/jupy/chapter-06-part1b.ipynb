Four main methods&mdash;for list, `L`:

- `L[a:b]` returns all items starting at index `a` until but __not__ including the item at index `b`
- `L[a:]` returns all items starting at index `a` until and including the last item in `L`
- `L[:b]` returns all items starting from the first until but __not__ including the item at index `b`
- `L[:]` returns a duplicate copy of `L`&ndash;equivalent to `list( L )`.

```python
A, B, C, D, E, F, G, H, I = range( 1, 10 )
```

```python
L = [A, B, C, D, E, F, G, H, I]
```

```python
L
```

```python
L[3:7] # D-->G
```

```python
L[:5] # A-->E
```

```python
L[5:] # F--I
```

Even if the end index is beyond the range of indices for `L`, it will still work.

```python
L[:15]
```

```python
L2 = L[:] # DUPLICATING, creating a physically separate copy
```

```python
L2 == L # Value equality
```

```python
L2 is L # But not memory equality!
```

Note that, again, the value at the end index is not included in the returned value.

With negative indices, we learn that the _left-to-right_ specification of indices in a _slice_ operation is not quite the way things happen. #### 6.3.3 Element Separators

When we display a list of values, we separate these values with a *delimiter*.

    Harry, Emily, Bob, Joe, Mary, Kate, Peter

Note that we need one less separator than there are  names. 

So we should add a delimiting character after the first value until but not including the last value.

```python
friends = ["Harry", "Emily", "Bob", "Joe", "Mary", "Kate", "Peter"]
delimiter = ", "
result = ""

for i in range( len( friends ) ):
    #
    if i > 0:
        #
        # If we are past the first item, we should add a delimiter
        # before adding the next item in the list
        #
        result += delimiter
        
    result += friends[i]
    
    print( f"--- i: {i}: result: \"{result}\"" ) # Print the running result
    
print( result )
```

We can print out the values on a single line with a separator in between each value.

```python
delimiter = " | "

for i in range( len( friends ) ): 
    #
    if i > 0 :
        print( delimiter, end="" ) 

    print( friends[i], end="" )
    
print()
```

```python
limit = 20 # Looking for a value larger than this value
pos = 0 # Starting index into the list
is_found = False # We have not yet found a value that is larger than limit
first_val = None # There is no value found yet to assign into this variable

# Uncomment so that the else part of the last if-statement will
# be forced to take effect
#
# values = range( 1, 20 )

while pos < len( values ) and not is_found:
    #
    cur_val = values[pos]
    
    if cur_val > limit:
        #
        first_val = cur_val
        is_found = True
    else:
        pos += 1

if is_found:
    print( f"--- The first value larger than {limit} is {first_val} (index: {pos})" )
else :
    print( f"### No value found larger than {limit}!" )
```

Note that, in the worst case, we will have to visit all elements in the input list!

Let us solve this problem using a for-loop instead of a while-loop.

```python
limit = 20 # Looking for a value larger than this value
first_val = None # There is no value found yet to assign into this variable

def check_larger_than( value, limit ):
    """
    Returns True if the provided value is larger than the limit
    """
    return value > limit

for cur_index, cur_value in enumerate( values ):
    #
    if check_larger_than( cur_value, limit ):
        #
        first_val = cur_value
        break

if first_val:
    #
    print( f"--- The first value larger than {limit} is {first_val} (index: {cur_index})" )
else :
    print( f"### No value found larger than {limit}!" )
```

#### 6.3.7 Removing Matches

Let's write code to remove all strings in a list with less than, say, four characters.

```python
words = "Welcome to the island of Cyprus!".split()
max_chars = 4

print( f"--- words: {words}" )

for i in range( len( words ) ):
    # 
    print( f"--- Accessing the word at index {i}", end="" )
    cur_word = words[i]
    print( f": '{cur_word}'" )

    if len( cur_word ) < max_chars:
        #
        # Remove the element at index i
        #
        print( f"\t<<< Removing word '{cur_word}' at index {i}" )
        removed = words.pop( i ) 
        print( f"\t/// Words: {words}" )
```

___Problem #1:___

> Note that the for-loop is initialized only once. That is the `range( len( words ) )` is only executed on entering the loop. Therefore the for-loop will want to iterate six times, since there are six words in the `words` list. <br><br>Since we progressively keep removing the elements that satisfy the condition from the `words` list, the list keeps getting shorter every time.<br><br>After removing `to` and `of`, we only have four words remaining in the list. Therefore, when the index becomes `4`, there is no fifth item in the list to access any longer.

___Problem #2:___

> Note that, after removing `to`, we ended up removing `of`, having skipped `the`, which also satisfies the condition of having less than four characters.<br><br>What happened?

    words = ["Welcome", "to",  "the", "island", "of", "Cyprus!"]
    indices:     0        1      2        3       4       5
    
> After removing the word `to` at position `1`, the list `words` will become

    words = ["Welcome", "the", "island", "of", "Cyprus!"]
    indices:     0        1      2        3       4
    
> and the index would be `2` when execution goes back up to the top of the for-loop. However, another word, `the`, which satisfies the removal condition will be skipped, leading to an error.

Here's one method for properly removing items that satisfy a given condition.

```python
words = "Welcome to the island of Cyprus!".split()
max_chars = 4
result = []

for cur_word in words:
    # 
    if len( cur_word ) >= max_chars:
        #
        # Remove the element at index i
        #
        result.append( cur_word )
result
```

Another method:

```python
words = "Welcome to the island of Cyprus!".split()
max_chars = 4
result = [cur_word for cur_word in words if len( cur_word ) >= max_chars]
result
```

Yet another method:

```python
result = filter( lambda v: len( v ) >= max_chars, words )
list( result )
```

#### 6.3.8 Swapping Elements

To swap the values of two elements in a list, `values` at indices `i` and `j`, we can employ a temporary variable, `temp`. First we save the value at index `i` to `temp`. Second we copy the value from index `j` to index `i`. 

    temp = values[i]
    values[i] = values[j]

Finally we set `values[j]` to the value in the temporary variable, which has the original value of `values[i]`:

    values[j] = temp

```python
values = [2, 6, 1, 8, 3, 7, 4, 9, 5]
```

All we need to do is to do is swap the indices such that `i` corresponds to `j`, and `j` corresponds to `i`: 

```python
i, j = 2, 7
values[i], values[j] = values[j], values[i]
```

Here's a function that returns the sum of a list of numbers.

```python
def sum_all( list_of_numbers ):
    """
    Returns the sum of the provided list of numbers.
    """
    total = 0 # INITIALIZE the sum to zero
    for cur_number in list_of_numbers:
        total += cur_number
    return total
```

```python
t = [1, 2, 3, 4, 5, 6]
sum_all( t )
```

```python
def multiply( values, factor ):
    """
    Multiples each value in the values list by the given factor. The
    original list will be modified, and this modification will be visible
    from the calling context even though this function does not and need not
    return the updated values list.
    
    The reason is that Python passes a reference to objects in calling functions.
    Therefore, for efficiency, Python does not make a duplicate of compound data
    types such as lists.
    """
    for i in range( len( values ) ):
        #
        values[i] *= factor
```

```python
values = list( range( 1, 11 ) )

print( f"--- values: {values}" )
multiply( values, 10 )
```

Let us write a function that will non-destructively reverse the order of the items in a given list. 

```python
def non_destructive_reverse_v1( old_lst ):
    """
    """
    new_lst = [] # INITIALIZATION!!!
    
    for cur_elem in old_lst:
        #
        # NOTE that appending will preserve the original order of elements.
        # So we cannot use append() to reverse the order of items. Instead
        # we will use insert() and insert every item at the beginning always.
        #
        new_lst.insert( 0, cur_elem )
    
    return new_lst
```

```python
lst = [1, 2, 3, 4, 5]
non_destructive_reverse_v1( lst )
```

We can write this function slightly differently using list concatenation making sure that we always add the next element from the list at the beginning of the new list that we are building.

```python
def non_destructive_reverse_v2( old_lst ):
    """
    """
    new_lst = [] # INITIALIZATION!!!
    
    for cur_elem in old_lst:
        #
        # Adding the new element to the beginning!
        # Also note that we have to create a one-element
        # list so that we can concatenate the current
        # element into new_lst.
        #
        new_lst = [cur_elem] + new_lst
        
    return new_lst
```

```python
lst = [11, 2, 3, 4, 55]
non_destructive_reverse_v2( lst )
```

```python
def argument_passing( param, mode, new_value, index=None ):
    """
    """
    print( f"(A) param: {param}, id: {id( param )}" )
    
    if mode == 'modify_param_value':
        #
        print( f"(B) Modying param {param} <-- new_value {new_value}" )
        param = new_value # Creates a local copy with the same name
    
    elif mode == 'modify_element_value':
        #
        print( f"(C) Modying param[index] {param[index]} <-- new_value {new_value}" )
        param[index] = new_value
        
    print( f"(D) param: {param}, id: {id( param )}" )
```

#### __Rolling the Dice: Fairness__

How can we decide if a die is _fair_?

Since a die has six sides, we would expect that we should get each side with equal probability, if we throw the die theoretically speaking an infinite number of times. 

However, simulating the rolling of a die an infinite number of times is obviously impossible. However, we can try throwing the die a relatively large number of times and compute the percentage with which we get 1, 2, 3, 4, 5, and 6.

If the probability of occurrence of each side is equal, that is, $\dfrac{1}{6}$ ideally, we know that have a *fair* die, since each side is expected to have an *equal* change of occurring.

<span style="color: #080; font-weight: 700;">Step 1: Decompose this task into steps</span>

__WHAT__&mdash;___NOT HOW___&mdash;will we have to do to accomplish this task?

- Need to throw the die (or simulate the throwing of the die) _many_ number of times
- After each throw we should record which side we got
- After many throws, we should tally how many times we got each side
- Compute the percentage of how many times we got each side
- Inspect whether each percentage is approximately equal to $\frac{1}{6}$.

<span style="color: #080; font-weight: 700;">Step 2: Determine which _algorithms_ we will need</span>

- Need to simulate the throwing of the die (`random.randint()`)
- Need to keep track of which side we got (`list`)
- Need to compute the _percentage_ of how many times we got each side

We do not really need very complicated algorithms in this case.,

<span style="color: #080; font-weight: 700;">Step 3: Use functions to structure your program</span>

- `def roll_die( roll_function )`
- `def make_trials( n )`
- `def compute_percentages( values )`
- `def print_statistics( counts, percentages )`
- `def check_fairness( percentages, precision )`

<span style="color: #080; font-weight: 700;">Step 4: Develop incrementally, integrate, and test your program</span>

```python
import random

def roll_die( roll_funct=lambda: random.randint( 1, 6 ) ):
    """
    Simulates the rolling of a die. It returns the value
    of the side in the current throw.
    """
    side = roll_funct()
    return side
```

```python
def make_trials( n, roll_funct=lambda: random.randint( 1, 6 ) ):
    """
    Rolls a die n number of times and records each side in a list
    that it returns after the end of this process.
    """
    rolls = [] # INITIALIZE
    
    for _ in range( n ):
        #
        rolls.append( roll_die( roll_funct=roll_funct ) )
    
    return rolls
```

```python
def make_trials__( n ):
    """
    Rolls a die n number of times and records each side in a list
    that it returns after the end of this process.
    """
    rolls = [roll_die() for _ in range( n )]
    return rolls
```

To compute the percentage of how many times each side occurred, we will need to think first:
    
Again, as you should always do, __SOLVE A PROBLEM ON PAPER FIRST!__

So suppose we get the kind of values we obtained from `make_trials()` in above cases.

Can we count how many 1s, 2s, and so on we got?

Note that we will need to have 6 percentage values at the end of this process, one for each side of the die.

So let's write an *algorithm*.

This algorithm looks just OK, but, it's not the best one we can come up with. So let's try a little harder, getting rid of all specific variables and all other unnecessary length complication.

I will not worry about testing this algorithm __ON PAPER__ myself. Please do it as an exercise on your own. This is a required step you must learn to go through!

```python
def compute_percentages( die_rolls ):
    """
    Computes how many times each value occurred. It returns both the
    count of each value and its percentage in two separate lists
    """
    counts = [0] * 6
    percentages = [None] * 6
    
    for cur_roll in die_rolls:
        #
        counts[cur_roll - 1] += 1
        
    total_throws = len( die_rolls )
    
    for count_idx, count_val in enumerate( counts ):
        #
        percentages[count_idx] = count_val * 100 / total_throws
        
    return counts, percentages
```

```python
def print_statistics( counts, percentages ):
    """
    Prints out how many times we rolled a given side of a die and with
    what percentage
    """
    for side_val in range( 1, 7 ):
        #
        side_cnt = counts[side_val - 1]
        side_pct = percentages[side_val - 1]
        
        print( f"--- Threw side {side_val} {side_cnt:10d} time(s) [{side_pct:4.1f}%]" )
    
    print()
```

Finally, we need to implement the last function that, given percentages of the rolls of each side will decide whether the die in question is fair or not.

Since we will not get perfect percentages, that is, $\frac{1}{6}$ each, we will have to accept percentages or _probabilities_ within a _margin of error_, similar to the way we were checking for the equality of two floating-point values.

```python
def check_fairness( percentages, precision=0.001, perfect_percentage=100.0/6.0 ):
    """
    """
    fairness = [None] * len( percentages )
    
    for pct_idx, pct_val in enumerate( percentages ):
        #
        abs_diff = abs( pct_val - perfect_percentage )
                
        if abs_diff > precision:
            #
            print( f"### UNFAIR side: {pct_idx + 1}" )
            fairness[pct_idx] = False
        else:
            fairness[pct_idx] = True
    
    return fairness
```

```python
def check_fairness( percentages, precision=0.001, perfect_percentage=100.0/6.0 ):
    """
    """
    fairness = [None] * len( percentages )
    
    for pct_idx, pct_val in enumerate( percentages ):
        #
        abs_diff = abs( pct_val - perfect_percentage )
                
        if abs_diff > precision:
            #
            print( f"### UNFAIR side: {pct_idx + 1}" )
            fairness[pct_idx] = False
        else:
            fairness[pct_idx] = True
    
    if all( fairness ):
        #
        print( "--- The die is FAIR" )
    else:
        print( "*** The is NOT FAIR!!!" )
    
    return fairness
```
